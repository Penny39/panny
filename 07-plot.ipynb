```python
import numpy as np  
import matplotlib.pyplot as plt  
  
x = np.linspace(0, 2 * np.pi, 100)  # 创建x轴数据  
y = np.sin(x)  # 创建y轴数据  
  
fig, ax = plt.subplots()  # 创建一个包含一个子图的画布和轴对象  
ax.plot(x, y)  # 在当前子图上绘制折线图  
  
plt.show()  # 显示图形
```

```python
x=[1,2,3,4]
y=[1,4,2,3]
fig,ax=plt.subplots()
ax.plot(x,y) 
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
plt.show()
```

```python
import matplotlib.pyplot as plt  
plt.rcParams["font.sans-serif"]=["SimHe"] #设置字体  
plt.rcParams["axes.unicode_minus"]=False #正常显示负号  
year = [2017, 2018, 2019, 2020]  
people =[20, 40,60,70]  
#生成图表  
plt.plot(year, people)  
plt.xlabel("年份") #x轴标签  
plt.ylabel("人口") #y轴标签  
plt.title('人口增长') #图表标题  
#设置纵坐标刻度  
plt.yticks([0, 20, 40, 60, 80]) #设置y轴刻度  
#设置填充选项：参数分别对应x轴数据，y轴数据，y轴填充起始值，填充颜色  
plt.fill_between(year, people, 20, color='blue') #填充颜色为绿色，起始值为40  
#显示图表  
plt.show()
```
